```python
def print_m_shape(height):
    if height < 4: # Minimum height for a clear 'M' pattern
        print("Height must be at least 4 for a visible 'M' shape.")
        return

    width = height * 2 - 1
    for row in range(height):
        for col in range(width):
            # Left vertical line
            if col == 0:
                print("*", end="")
            # Right vertical line
            elif col == width - 1:
                print("*", end="")
            # Top left diagonal part
            elif row == col and row <= height // 2:
                print("*", end="")
            # Top right diagonal part
            elif row + col == width - 1 and row <= height // 2:
                print("*", end="")
            else:
                print(" ", end="")
        print()

# You can change the height to adjust the size of the 'M'
print_m_shape(7)
```
